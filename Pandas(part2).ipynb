```python
# Checking NULL/NaN Values in Pandas
# Null value means the value is missing! Not 0!!!
```

```python
data = np.random.rand(10, 3)
for col in range(data.shape[1]):
    null_indices = np.random.choice(10, 3, replace=False)
    data[null_indices, col] = np.nan
df = pd.DataFrame(data, columns=['Column_1', 'Column_2', 'Column_3'])
```

```python
df
```

```python
# Checking with isna()
pd.isna(df['Column_1']) # NaN value thakle true ashbe. naile False ashbe
```

```python
pd.isna(df) # Full dataframe er jonno
```

```python
# Checking with notna()
pd.notna(df['Column_1']) # opposite function to isna()
```

```python
pd.notna(df)
```

```python
df.isnull() # null value check korar arekta way. Function similar as isna()
```

```python
# Checking with isnull().sum()  MOSTLY USED!!! .info() use koreo null value koyta ache dekha jay
df.isnull().sum() # protita column a jotota NULL value ase ogular sum
```

```python
df.notnull().sum() # protita column a jotota VALUE ase ogular sum 
```
